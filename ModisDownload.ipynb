```python
class_list= ['RMG','GBG', 'SWG'
             #, 'SWC', 'SWD', 'SWG', 
            # 'SAA', 'SAB', 'SAC', 'SAD', 'SAG'
            ]

_site_array1 = [class_list[0] + str(x) for x in range(1, 200)]
_site_array2 = [class_list[1] + str(x) for x in range(1, 200)]
_site_array3 = [class_list[2] + str(x) for x in range(1, 200)]
'''
_site_array4 = [class_list[3] + str(x) for x in range(1, 200)]
_site_array5 = [class_list[4] + str(x) for x in range(1, 200)]
_site_array6 = [class_list[5] + str(x) for x in range(1, 200)]
_site_array7 = [class_list[6] + str(x) for x in range(1, 200)]
_site_array8 = [class_list[7] + str(x) for x in range(1, 200)]
_site_array9 = [class_list[8] + str(x) for x in range(1, 200)]
_site_array10 = [class_list[9] + str(x) for x in range(1, 200)]
'''

_site_array =  _site_array1 + _site_array2  + _site_array3  #+ _site_array4 + _site_array5 + _site_array6 + _site_array7 + _site_array8 + _site_array9 + _site_array10
_sites = dict.fromkeys(_site_array, {})  #[0:3]
_sites
```

```python
_rootdir = "Data/"
_products = ['MOD13Q1','MOD15A2H','MYD11A2']  #'MOD16A2', 
print(_products)
number_of_nulls = 0
_pixels = 0
for _product in _products:
    #print(_product)
    _product_dir = _rootdir+_product
    for _class in class_list:
        _class_dir = _product_dir+"/"+_class
        _files = os.listdir(_class_dir)
        _files.sort()
        for _file in _files:  
                
            _site, _band = fires.get_bandname_from_file(_file)
            #print(_file, _site, _band)
            
            _ds, _nulls, _size = fires.read_modisdata_file(_class_dir+"/", _file, _band)
            
            
            _d = _sites.get(_site).copy()  #Had to do this .... as dictionary was getting copied across all the sites
            
            _d[_band]=_ds
            _sites[_site]=_d
            number_of_nulls = number_of_nulls + _nulls
            _pixels = _pixels+ _size
            
        #break
#print(_ds)

print(number_of_nulls, _pixels)           
_dfResult = fires.convert_dictionary_to_dataframe(_sites)        

#_dfResult.to_csv("bioatmosphere_data.csv")
```
